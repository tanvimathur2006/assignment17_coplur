```python
import os
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Initialize Spark Session configured for local execution
spark = SparkSession.builder \
    .appName("SalesDataAnalysis") \
    .master("local[*]") \
    .getOrCreate()

# Reduce logging verbosity to clean up the output windows
spark.sparkContext.setLogLevel("ERROR")
print("Spark Session successfully initialized!")
```

```python
# 1. Read the CSV file into a DataFrame
csv_path = "sales.csv"
print(f"Reading data from {csv_path}...\n")
df = spark.read.csv(csv_path, header=True, inferSchema=True)
```

```python
# 2. Sort all products by sales in descending order and display
print("--- All Products Sorted by Sales (Descending) ---")
sorted_df = df.orderBy(col("sales").desc())
sorted_df.show()
```

```python
# 3. Display the top 3 products with the highest sales values
print("--- Top 3 Products with Highest Sales ---")
top_3_df = sorted_df.limit(3)
top_3_df.show()
```

```python
# 4. Filter products with sales greater than 80,000
print("--- Filtering Products with Sales > 80,000 ---")
filtered_df = df.filter(col("sales") > 80000)
filtered_df.show()
```

```python
# 5. Save the output as a CSV file
output_dir = "high_sales_output"
print(f"Saving filtered results to directory: {output_dir}...")
# Coalesce to 1 partition so it outputs as a single clean CSV file
filtered_df.coalesce(1).write \
        .mode("overwrite") \
        .option("header", "true") \
        .csv(output_dir)
```
